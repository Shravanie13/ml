```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dummies=pd.get_dummies(df.town) #encoding converting categories to numerical data
dummies
```

```python
merged=pd.concat([df,dummies],axis="columns")  #inplace true=1 can be used as one more that nmainmg the one in the variable
```

```python
merged
```

```python
merged=merged.drop(["town"],axis="columns")
```

```python
merged
```

When one variable can be predicted using other variables, it is called multicollinearity.

For example, if you know the values of California and Georgia, you can easily determine the value of New Jersey. This means these state variables are dependent on each other.

In such cases, linear regression won’t work correctly because the model struggles with redundant information. To fix this, you should remove one of the columns to avoid issues.


```python
merged=merged.drop(["west windsor"],axis="columns")
```

```python
merged
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
x = merged.drop(["price"],axis="columns")             #input features data
```

```python
y = merged.price #output variable
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.3, random_state = 42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
model=LinearRegression()
```

```python
model.fit(x_train,y_train)       #model.fit or fit method is used to train the model.
```

```python
y_test_pred=model.predict(x_test)
```

```python
y_test
```

```python
y_test_pred
```

```python
model.score(x_test,y_test)
```

```python
model.score(x_train,y_train)
```
